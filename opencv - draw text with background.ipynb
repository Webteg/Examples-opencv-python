Draw text with background using opencv in python this code was based on https://gist.github.com/aplz/fd34707deffb208f367808aade7e5d5c

```python
import cv2 # opencv
import matplotlib.pyplot as plt
%matplotlib inline

def put_text(text,text_offset_x,text_offset_y,img):
    # get the width and height of the text box
    rectangle_bgr = (255,255,0)
    font_scale = 1.5
    font = cv2.FONT_HERSHEY_DUPLEX
    (text_width, text_height) = cv2.getTextSize(text, font, fontScale=font_scale, thickness=1)[0]
    # set the text start position

    # make the coords of the box with a small padding of two pixels
    pad=5
    box_coords = ((text_offset_x-pad, text_offset_y+pad), (text_offset_x + text_width + pad, text_offset_y - text_height - pad))
    cv2.rectangle(img, box_coords[0], box_coords[1], rectangle_bgr, cv2.FILLED)
    cv2.putText(img, text, (text_offset_x, text_offset_y), font, fontScale=font_scale, color=(0, 0, 255), thickness=2)

if __name__ == "__main__":
    img=cv2.imread('images/lena.png')
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # # set some text
    text = "LENA"
    # # position 
    text_offset_x = 380
    text_offset_y = 250
    put_text(text,text_offset_x,text_offset_y,img_rgb)
    
    plt.figure(figsize=(10, 7))
    plt.imshow(img_rgb)
    plt.show()
```
